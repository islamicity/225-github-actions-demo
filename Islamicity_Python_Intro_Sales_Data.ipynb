```python
# Aggregate the publishers
aggregated_publishers = df['Publisher'].value_counts().to_dict()

# Find the publisher with the most games
most_games_publisher = max(aggregated_publishers, key=aggregated_publishers.get)
most_games_count = aggregated_publishers[most_games_publisher]

# Print the publisher with the most games
print(f"The publisher with the most games is {most_games_publisher} with {most_games_count} games.")
```

```python
# Convert the dictionary to a DataFrame
df_publisher = pd.DataFrame.from_dict(aggregated_publishers, orient='index', columns=['Name'])
df_publisher = df_publisher.sort_values('Name', ascending=False).head(10)

# Set the figure size
plt.rcParams['figure.figsize'] = [25, 10]

# Plot the bar chart
sns.barplot(x=df_publisher.index, y='Name', data=df_publisher)

# Customize the chart
plt.xlabel('publisher')
plt.ylabel('Game Count')
plt.title('Top 10 publishers by Game Count')

# Display the chart
plt.show()
```

```python

# Aggregate the publishers
aggregated_publishers = df.groupby('Publisher')['Name'].apply(list).to_dict()

# Print the aggregated publishers
for Publisher, Name in aggregated_publishers.items():
    print(f"Publisher: {Publisher}")
    print("Name:")
    for Name in Name:
        print(f"- {Name}")
    print()
```

```python
# Aggregate the developers
aggregated_developers = df['Developer'].value_counts().to_dict()

# Find the developer with the most games
most_games_developer = max(aggregated_developers, key=aggregated_developers.get)
most_games_count = aggregated_developers[most_games_developer]

# Print the developer with the most games
print(f"The developer with the most games is {most_games_developer} with {most_games_count} games.")
```

```python
df_cleaned = df.copy()
df_cleaned['Date'] = pd.to_datetime(df_cleaned['Date'])
df_cleaned['Revenue'] = df_cleaned['Unit price'] * df_cleaned['Quantity']

print(df_cleaned.dtypes)
df_cleaned.head()
```

```python
# check typo

df_cleaned['City'].value_counts()
```

```python
# check typo

for col_name in ['Payment','Branch', 'City', 'Customer type',
       'Gender', 'Product line' ]:
       print(df_cleaned[col_name].value_counts(),'\n')
```

```python
dict_typo = {'male': 'Male',
             'female': 'Female',
             'fml' : 'Female'}

df_replaced = df_cleaned.replace(dict_typo)
df_replaced.head()
```

```python
#check is there any duplicated invoice
df_replaced['Invoice ID'].duplicated().sum()
```

```python
#re-checking
for col_name in ['Payment','Branch', 'City', 'Customer type',
       'Gender', 'Product line' ]:
       print(df_replaced[col_name].value_counts(),'\n')
```

### Daily Trend Overall

```python
agg_daily = df_replaced.groupby('Date', as_index=False)['Revenue','Quantity'].sum()
agg_daily.head()
```

```python
plt.rcParams["figure.figsize"] = (20,5)

plt.plot('Date','Revenue', data=agg_daily)

plt.title('Revenue trend')
plt.xlabel('Date')
plt.ylabel('Rev($)')
plt.show()
```

### By Payment

```python
agg_payment = df_replaced.groupby('Payment', as_index=False).agg({'Revenue':'sum', 'Invoice ID': 'nunique'})
agg_payment
```

```python
plt.rcParams["figure.figsize"] = (10,5)
payment_data = agg_payment.sort_values('Revenue', ascending=False)
sns.barplot(x='Payment', y='Revenue', data=payment_data)
```

```python
plt.rcParams["figure.figsize"] = (10,5)
payment_sorted_data = agg_payment.sort_values('Invoice ID', ascending=False)
sns.barplot(x='Payment', y='Invoice ID', data=payment_sorted_data)
```

```python
plt.rcParams["figure.figsize"] = (10,5)
sns.barplot(x='Payment', y='Revenue', data = agg_payment.sort_values('Revenue', ascending=False))
```
